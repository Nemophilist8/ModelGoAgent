```python
from dotenv import load_dotenv
load_dotenv()
from e2b_code_interpreter import Sandbox
```

```python
sbx = Sandbox.create('model-go') # By default the sandbox is alive for 5 minutes
```

```python
execution = sbx.run_code("print('hello world')") # Execute Python inside the sandbox
print(execution.logs)
```

```python
sbx.files.list("/home/user/scripts")
```

```python
sbx.commands.run(
    'cd /home/user/scripts && python -c "'
    'from main_case import *; '
    'case1([deep_sequoia_text, pubmed_text], bigtranslate_model, [bigtranslate_model])'
    '"'
)
```

```python
SEPARATOR = "---------------------------------------------------------------------------------------------"
code_list = [
    "arxiv_text = Work('arXiv', 'data', 'raw', 'CC-BY-NC-SA-4.0') # Corpus",
    "stack_exchange_text = Work('StackExchange', 'data', 'raw', 'CC-BY-SA-4.0') # Corpus",
    "bigtranslate_model = Work('BigTranslate', 'model', 'raw', 'GPL-3.0') # Text Translation",
    "deep_sequoia_text = Work('Deep-sequoia', 'data', 'raw', 'LGPL-LR') # Corpus",
    "free_law_text = Work('FreeLaw', 'data', 'raw', 'CC-BY-ND-4.0') # Corpus",
    "works = [ob for ob in gc.get_objects() if isinstance(ob, Work)]",
    "par.register_license(works)",
    '\ntranslated = [embed(arxiv_text, bigtranslate_model), embed(stack_exchange_text, bigtranslate_model)]\nnew_work = combine(translated + [deep_sequoia_text, free_law_text])\n',
    "new_work.form = 'binary'", "par.analysis(new_work, open_policy='sell')",
    "new_work.summary()",
    f"print('\\n{SEPARATOR}\\n')",
    "print(new_work)"
]
```

```python
code ='from main_case import *\n'
code += "\n".join(code_list)
with open("../../scripts/tmp_run.py", "w") as f:
    f.write(code)
```

```python
sbx.files.write('/home/user/scripts/tmp.py', code)
result = sbx.commands.run(
    "cd /home/user/scripts && python tmp.py"
)
result.stdout
```
